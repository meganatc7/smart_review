```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import requests as req
```

```python
# 웹 브라우저 실행
driver = wb.Chrome('chrome/chromedriver.exe')

# 실행해서 지정한 url주소로 이동
url = 'https://naver.com'
driver.get(url)
```

```python
# 네이버 검색창 태그 검색

#1. xpath
input_search_x = driver.find_element_by_xpath('//*[@id="query"]')
input_search_x.send_keys('광주날씨')

#2.id
#input_search_id = driver.find_element_by_id('query')
#input_search_id.send_keys('광주날씨')
```

```python
# 엔터
input_search_x.send_keys(Keys.ENTER)
```

```python
# 이전 페이지
driver.forward()
driver.back()
```

```python
# 검색창 클릭
# 검색버튼 태그 검색
btn_search = driver.find_element_by_xpath('//*[@id="query"]')

# 검색버튼 클릭
btn_search.click()
```

```python
# 스크롤링
body = driver.find_element_by_tag_name('body')
for i in range(3):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

# 한솥 도시락 메뉴 이름, 가격 정보 수집

## 한솥 메뉴 페이지 들어가기

```python
url = 'https://www.hsd.co.kr/menu/menu_list'
driver = wb.Chrome('chrome/chromedriver.exe')

driver.get(url)
```

## 더보기 버튼 클릭하기

```python
btn_more = driver.find_element_by_xpath('//*[@id="btn_more"]/span/a')

# 더보기 버튼 3번 클릭
for index in range(3):
    btn_more.click()
    time.sleep(2)
```

## 도시락 정보 가져오기

```python
soup = bs(driver.page_source,'html.parser')

num = [] #메뉴 번호
names = [] # 메뉴 이름
prices = [] # 메뉴 가격

# 메뉴 번호와 메뉴 가격 태그 가져오기
name = soup.select('.item-text > h4')
price = soup.select('.item-price > strong')

# 반복문으로 리스트에 넣어줌
for idx in range(len(name)):
    num.append(idx+1)
    names.append(name[idx].text)
    prices.append(price[idx].text)

# 길이 맞게 잘 가져왔는지 확인
print(len(num), len(prices), len(names))
```

## 가져온 도시락 정보를 리스트에 담기

```python
lunchBox_info = {
    '메뉴 번호': num,
    '메뉴 이름': names,
    '메뉴 가격': prices
}
lunchBox = pd.DataFrame(lunchBox_info)
lunchBox.set_index('메뉴 번호',inplace=True)
lunchBox.head()
```

## csv파일로 저장

```python
lunchBox.to_csv('data/hansot_menu.csv',encoding='utf-8')
```

# 스타벅스 매장 지점명, 주소, 번호 정보 수집

## 스벅 접속

```python
url = 'http://www.istarbucks.co.kr/store/store_map.do'
driver = wb.Chrome('chrome/chromedriver.exe')
driver.get(url)
```

## 지역검색

```python
# 이벤트 창 닫기
btn_x = driver.find_element_by_css_selector('.holiday_notice_close > a')
btn_x.click()
# 지역검색
btn_search = driver.find_element_by_xpath('//*[@id="container"]/div/form/fieldset/div/section/article[1]/article/header[2]/h3/a')
btn_search.click()

# 광주지역
btn_gwangju = driver.find_element_by_xpath('//*[@id="container"]/div/form/fieldset/div/section/article[1]/article/article[2]/div[1]/div[2]/ul/li[3]/a')
btn_gwangju.click()

time.sleep(1)
# 광주 전체 클릭
allArea = driver.find_element_by_xpath('//*[@id="mCSB_2_container"]/ul/li[1]/a')
#allArea = driver.find_element_by_css_selector('ul.gugun_arae_box > li')
allArea.click()
```

## 정보 가져오기

```python
soup = bs(driver.page_source,'html.parser')

name_list = []
addr_list = []
tel_list = []

names = soup.select('.quickResultLstCon > strong')
addrs = soup.select('.result_details')

for i in range(len(names)):
    name_list.append(names[i].text)
    addr_list.append(addrs[i].text[:-9])
    tel_list.append(addrs[i].text[-9:])
    
print(name_list[0], addr_list[0], tel_list[0])
```

```python
starbucks_info = {
    '지점': name_list,
    '주소': addr_list,
    '번호': tel_list
}
starbucks = pd.DataFrame(starbucks_info)
starbucks.head()
```

# 스타벅스 음료 정보 가져오기

## 스타벅스 메뉴 페이지 접속

```python
url = 'http://www.istarbucks.co.kr/menu/index.do'
driver = wb.Chrome('chrome/chromedriver.exe')
driver.get(url)
```

## 음료 정보 보기

```python
# 자세히보기
driver.find_element_by_xpath('//*[@id="container"]/div[6]/div/div[2]/a').click()
# 영양정보 보기
driver.find_element_by_xpath('//*[@id="container"]/div[2]/div[2]/div/dl/dt[2]/a').click()
```

## 음료 정보 수집

```python
soup = bs(driver.page_source,'html.parser')

drink_cols = soup.select('table.coffeeInfo.mb60 th',limit=7)
drink_info = soup.select('table.coffeeInfo.mb60 td')
drink_info
```

```python
col_list = []
info_list = []

# 각각 가져온 정보의 text만 리스트에 저장
for i in drink_info:
    info_list.append(i.text)
    
for i in drink_cols:
    col_list.append(i.text)
```

```python
# 음료 정보를 데이터프레임으로 만들기 위해 2차원으로 만들기
info_list_2 = []

# 음료 영양분 정보가 7개니까 7개씩 리스트로 담아서 2차원으로
for i in range(len(info_list)//7):
    info_list_2.append(info_list[i*7:i*7+7]) 
```

```python
starbucks_beverage = pd.DataFrame(info_list_2, columns=col_list)
starbucks_beverage.set_index('메뉴', inplace=True)
starbucks_beverage
```
