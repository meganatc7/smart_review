### 다차원 배열을 효과적으로 처리할 수 있도록 도와주는 도구

list보다 빠르고 다양한 기능을 제공
- 1차원 데이터(=벡터)
- 2차원 데이터(=매트릭스)
- 3차원 이상 데이터(=텐서, 채널)

np.array를 사용하여 빠른 연산이 가능!!

```python
import numpy as np
```

```python
# numpy를 사용한 1차원 배열

array = np.array([1,2,3,4])
print(array)
```

```python
# numpy를 사용한 2차원 배열

array_2 = np.array([[1,2,3],[4,5,6]])
print(array_2)
```

```python
# 배열 크기 shape

print(array.shape)
print(array_2.shape) # 2행 3열
```

```python
# 배열 전체의 요소 개수 확인

print(array_2.size)
print(array.size)
```

```python
# 배열의 차원 확인 ndim

print(array_2.ndim)
print(array.ndim)
```

### 배열 생성하기

```python
arr_zero = np.zeros((3,4)) # 소괄호로 행과 열 지정
print(arr_zero)
```

```python
# 1값으로 모두 초기화

arr_one = np.ones((3,4))
print(arr_one)
```

```python
# 임의의 값으로 모두 초기화

arr_f = np.full((5,5),"f")
print(arr_f)
```

```python
# 1~50이 담긴 배열

list1 = []
for i in range(1,51):
    list1.append(i)

array_3 = np.array(list1)
print(array_3)
```

```python
# 1~50값 입력 arange(시작값, 종료값)

array_4 = np.arange(1,51)
print(array_4)
```

```python
# 랜덤 값 배열 생성

arr_ran = np.random.rand(2,3)
print(arr_ran)
```

```python
# array 연산

array_5 = np.array([1,2,3])

print(array_5 + array_5)
print(array_5/2.0)
print(array_5*3)
```

```python
with open("height_weight.txt",'r')as f:
    height = f.readline()
    weight = f.readline()
    
print(height) # 175.2,180.3!!!가 하나의 큰 덩어리 데이터
print(weight) # 마찬가지

height = height.split(",")
weight = weight.split(",")
print(height)
print(weight)

#문자열을 숫자로 형변환하는 방법. dtype은 np.array의 타입을 변경
height_np = np.array(height, dtype = np.float64)
weight_np = np.array(weight, dtype = np.float64)

print(height_np)
print(weight_np)

# BMP = 몸무케/키(m) x 키(m), cm인 키를 m로 환산해줘야함
height_np /= 100 
print(height_np)
bmi = weight_np / (height_np*height_np)
print(bmi)
```

```python
# 인덱싱

arr1 = np.arange(10)
print(arr1)
print(arr1[0])
print(arr1[-1])
```

```python
# 슬라이싱

print(arr1[2:5])
print(arr1[2:])
print(arr1[:5])
print(arr1[:])
```

```python
# reshape(행,렬) 2차원 배열 정의

arr2 = np.arange(50).reshape(5,10)
print(arr2)
```

```python
print(arr2[2:4]) # 2~4번째 미만의 행 출력
print("=====================")
print(arr2[2:]) # 2번째 행 부터 끝까지
print("=====================")
print(arr2[:4]) # 처음부터 4번째 미만 행 까지
print("====================")
print(arr2[:]) # 전체 출력
```

```python
print(arr2[:,:]) # 쉼표를 기준으로 행과 열로 나누어짐
print("==============================================")
print(arr2[:2,:]) # 2번째 행 미만 값을 모두 출력
print("==============================================")
print(arr2[:,0]) # 모든 행의 0번째 열만 출력
print("==============================================")
print(arr2[:4,:5]) # 0~4번째 미만의 행에서 0~5번째 미만의 열까지 출력
print("==============================================")
print(arr2[2,1]) # 2행 1열에 있는 데이터 출력
```

```python
# boolean 인덱싱 - True값만 출력 해 주는 것 (데이터 시각화)

name = np.array(["1호","2호","3호","4호"])
bol = np.array([True, True, False, False])

print(name[bol])

print(name[name=="1호"]) # True인 데이터만 출력
```

```python
arr3 = np.array([11,22,33,44,55])
print(arr3[arr3>30]) # True인 데이터만 출력
```

```python
# 단일 배열에서 사용되는 함수
arr5 = np.random.rand(5,4)
print(arr5)

print(arr5.sum()) # arr5에 있는 모든 수 더하기
print(arr5.mean()) # 평균 구하기
```

```python
arr6 = np.arange(1,6)
print(arr6)
print(np.sqrt(arr6)) # 제곱근
```

```python
arr7 = np.array([-1,5,-11,7,-20])
print(np.abs(arr7)) # 절대값
```

### Numpy를 활용한 영화 리뷰 분석

```python
data = np.loadtxt("ratings_small.dat", delimiter = '::', dtype = np.int64)
print(data)
```

```python
print(data.ndim) # 몇 차원 배열인지
print(data.shape) # 몇행 몇열인지
print(data.size) # 배열 전체 요소 개수 확인
```

```python
data[:,2] # 영화 평점인 모든행의 2열을 출력
```

```python
# 영화 평점의 평균

all_mean = data[:,2].mean()
print(all_mean)
```

```python
# 중복 값 제거. unique로 유일한 값만 출력 id의 중복된 것들을 제거함
user_id = np.unique(data[:,0])
print(user_id)
print(user_id.size)
print(user_id.shape) # 1779행
```

```python
# 사용자 아이디가 1인 사람의 데이터만 boolean 인덱싱
data_1 = data[data[:,0]==1]
data_2 = data[data[:,0]==2]
data_3 = data[data[:,0]==3]
data_4 = data[data[:,0]==4]
data_5 = data[data[:,0]==5]
```

```python
print(data_1.shape) # 53행, 즉 아이디 1이 53개의 리뷰 남김
```

```python
data_1[:,2] # 1번 사람이 매긴 평점 슬라이싱
```

```python
data_avg1 = data_1[:,2].mean() # 1번 사람이 남긴 평점의 평균
data_avg2 = data_2[:,2].mean()
data_avg3 = data_3[:,2].mean()
data_avg4 = data_4[:,2].mean()
data_avg5 = data_5[:,2].mean()
```

```python
# 전체 데이터 아이디별 평균 평점 총 1779개

cnt = 0

for i in user_id:
    data_rating = data[data[:,0]==i]
    data_rating_mean = data_rating[:,2].mean()
    print(data_rating_mean)
```

```python
# 각 유저가 매긴 평균평점ㄴ

data_usra = []

for i in user_id:
    data_rating = data[data[:,0]==i]
    data_rating_mean = data_rating[:,2].mean()
    data_usra.append([i,data_rating_mean])

print(data_usra)
```
